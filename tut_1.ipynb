```python
#Strings
ML="I am interested in machine learning"
print(ML)

#join elements in a string
print("--".join(ML))

#reversing a string
print("".join(reversed(ML)))

#slicing a string
print(ML[5:12])

#splitting a string
print(ML.split())

#change to lower case and upper case
print(ML.lower())
print(ML.upper())
```

```python
#LIST
stuff = ['laptop','notebook','choclate','bag']
print(stuff)

#length of a list
print(len(stuff))

#adding elements to list
stuff.append("joyboy")
print(stuff)

#reversing a list
stuff.reverse()
print(stuff)

#removing elements from list
stuff.remove('bag')
print(stuff)
```

```python
#Tuples
anime1=['jjk','op','HxH']
anime2=['BC','yN','Suzume']

#concatenation
anime3=anime1+anime2
print(anime3)

#slicing
print(anime3[1:])#removes first element
print(anime3[1:3])#prints element at index 1 and 2

#count the number of elements
print(anime3.count('HxH'))

#sorting
print(sorted(anime3))
```

```python
#SETS
random=set(['a',1,'bc','a',232,423])

#adding elements to a set
random.add('jjk')
print(random)

#removing a element from set
random.remove('jjk')
print(random)

#clear th set
random.clear()
print(random)

#difference
set1=set([1,2,3,4,5,6])
set2=set([4,5,6,7,8,9])
print(set1-set2)
print(set2-set1)
```

```python
#DICTIONARY
CHAR={'GOJO':'JJK','GON':'HxH','LUFFY':'ONE_PIECE','ASTA':'BLACK_CLOVER','naruto':'shipuden'}
print(CHAR)

#accessing elements
print(CHAR['GOJO'])

#DELETING
del(CHAR['GON'])
print(CHAR)

#ADDING ELEMENTS
CHAR.update({'GOKU':'DBZ'})
print(CHAR)
```

Given a list having Names, work_hours, and gender, Write a function to print name of the female worker that worked the most hours. Also how much do should she be paid if the pay is $ 20 per hour.
work_hours = [('Abby',100 , 'F'),('Billy',400, 'M'),('Cassie',800,'F'), ('Maggi',600,'F'),('Alex',500,'M'),('Raj',225,'M'),('Penny',920,'F'),('Ben',300,'M')]

```python
def work(a):
  salary =0
  highest = 0
  for i in range(0,7):
    if a[i][2]=='F':
      if a[i][1]>salary:
        salary = a[i][1]
        highest = i
  return highest
work_hours = [('Abby',100 , 'F'),('Billy',400, 'M'),('Cassie',800,'F'), ('Maggi',600,'F'),('Alex',500,'M'),('Raj',225,'M'),('Penny',920,'F'),('Ben',300,'M')]
highest_salary=work(work_hours)
most_worked = work_hours[highest_salary][0]
pay=(work_hours[highest_salary][1])*20
print(f"The female worker that worked the most hours is {most_worked} and she should be paid {pay}")
```
